# Import the necessary libraries for Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the Restaurant dataset into a DataFrame

```python
df=pd.read_csv('dataset.csv',encoding='ISO-8859-1')
df.head()
```

# LEVEL 2-Task 3: Feature Engineering

# LEVEL 2-Task 3(i):Extract additional features from the existing columns, such as the length of the restaurant name or address.

```python
df['Restaurant Name Length']=df['Restaurant Name'].apply(len)
df['Address Length']=df['Address'].apply(len)
df['Cuisines Count']=df['Cuisines'].apply(lambda x:len(x.split(','))if pd.notna(x) else 0)
df['Has Menu']=df['Switch to order menu'].apply(lambda x: 1 if x.lower()=='yes' else 0)
print(df)
```

# LEVEL 2-Task 3(ii):Create new features like "Has Table Booking" or "Has Online Delivery" by encoding categorical variables.


```python
df['Has Table Booking']=df['Has Table booking'].map({'Yes':1,'No':0})
df['Has Online Delivery']=df['Has Online delivery'].map({'Yes':1,'No':0})
print(df)
```
