# Import the necessary libraries for Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Loading the Restaurant dataset into a DataFrame

```python
df=pd.read_csv('dataset.csv',encoding='ISO-8859-1')
df.head()
```

# LEVEL 1-Task 2: Descriptive Analysis

# LEVEL 1-Task 2 (i):Calculate basic statistical measures (mean,median, standard deviation, etc.) for numericalcolumns.

```python
df.describe()
```

```python
numeric_columns=df.select_dtypes(include=['int64','float64'])
mean=numeric_columns.mean()
median=numeric_columns.median()
std_dev=numeric_columns.std()
minvalue=numeric_columns.min()
q1=numeric_columns.quantile(0.25)
q2=numeric_columns.median()
q3=numeric_columns.quantile(0.75)
maxvalue=numeric_columns.max()
print("Mean Values:")
print(mean)
print("\nMedian Values:")
print(median)
print("\nStandard Deviation Values:")
print(std_dev)
print("\nMinimum Values:")
print(minvalue)
print("\n25th Percentile(Q1)Values:")
print(q1)
print("\n50th Percentile(Q2 or Median) Values:")
print(q2)
print("\n75th Percentile(Q3) Values:")
print(q3)
print("\nMaximum Values:")
print(maxvalue)
```

# LEVEL 1-Task 2 (ii):Explore the distribution of categorical variables like "Country Code", "City," and"Cuisines".

```python
country_code_distribution=df['Country Code'].value_counts()
print("Distribution of Country Code:")
print(country_code_distribution)

city_distribution=df['City'].value_counts()
print("\nDistribution of City:")
print(city_distribution)

cuisines_distribution=df['Cuisines'].value_counts()
print("\nDistribution of Cuisines:")
print(cuisines_distribution)

top_country_codes=country_code_distribution.head(3)
top_country_codes.plot(kind='bar',title='Top 3 Country Codes')
plt.xlabel('Country Code')
plt.ylabel('Count')
plt.show()

top_cities=city_distribution.head(3)
top_cities.plot(kind='bar',title='Top 3 Cities')
plt.xticks(rotation=45)
plt.xlabel('City')
plt.ylabel('Count')
plt.show()

top_cuisines=cuisines_distribution.head(3)
top_cuisines.plot(kind='pie',title='Top 3 Cuisines',labels=top_cuisines.index,autopct='%1.1f%%',startangle=90)
plt.axis('equal')
plt.show()
```

# LEVEL 1-Task 2 (iii):Identify the top cuisines and cities with the highest number of restaurants.

```python
top_cuisines = df.groupby('Cuisines')['Restaurant Name'].count().sort_values(ascending=False).head(3)
print("Top Cuisines:")
print(top_cuisines)

top_cities = df.groupby('City')['Restaurant Name'].count().sort_values(ascending=False).head(3)
print("\nTop Cities:")
print(top_cities)
top_cuisines.plot(kind='bar', title='Top Cuisines')
plt.xticks(rotation=45)
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.show()

top_cities.plot(kind='bar', title='Top Cities')
plt.xticks(rotation=45)
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```
