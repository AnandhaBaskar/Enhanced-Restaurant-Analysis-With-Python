# Import the necessary libraries for Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Loading the Restaurant dataset into a DataFrame

```python
df=pd.read_csv('dataset.csv',encoding='ISO-8859-1')
df.head()
```

# LEVEL 2-Task 1 :Table Booking and Online Delivery

# LEVEL 2-Task 1 (i):Determine the percentage of restaurants that offer table booking and online delivery.

```python
df['Has Table booking']=df['Has Table booking'].map({'Yes':True,'No':False})
df['Has Online delivery']=df['Has Online delivery'].map({'Yes':True,'No':False})
table_booking_percentage=(df['Has Table booking'].sum()/df.shape[0])*100
online_delivery_percentage=(df['Has Online delivery'].sum()/df.shape[0])*100
print(f"Percentage of Restaurants with Table Booking:{table_booking_percentage:.2f}%")
print(f"Percentage of Restaurants with Online Delivery:{online_delivery_percentage:.2f}%")
```

# LEVEL 2-Task 1 (ii):Compare the average ratings of restaurants with table booking and those without.

```python
df_filtered = df.dropna(subset=['Has Table booking'])
average_rating_with_booking=df_filtered[df_filtered['Has Table booking']]['Aggregate rating'].mean()
average_rating_without_booking=df_filtered[~df_filtered['Has Table booking']]['Aggregate rating'].mean()
print(f"Average Rating with Table Booking:{average_rating_with_booking:.2f}")
print(f"Average Rating without Table Booking:{average_rating_without_booking:.2f}")
```

# LEVEL 2-Task 1 (iii):Analyze the availability of online delivery among restaurants with different price ranges.

```python
delivery_by_price=df.groupby('Price range')['Has Online delivery'].mean()*100
print("Percentage of Restaurants with Online Delivery by Price Range:")
print(delivery_by_price)
plt.figure(figsize=(10, 6))
delivery_by_price.plot(kind='bar',color='yellow')
plt.title('Percentage of Restaurants with Online Delivery by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Percentage with Online Delivery')
plt.xticks(rotation=0)
plt.show()
```
